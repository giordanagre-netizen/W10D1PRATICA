Cosa potremmo fare con la Series che non possiamo fare con il dizionario?

Abbiamo:
fatturati_dict[1997]
fatturati_series[1997]
Entrambi restituiscono 12_000.

e per accedere ai dati funzionano allo stesso modo.

Ma la somiglianza finisce qui.

Differenza fondamentale: DIZIONARIO vs SERIES
1-Il dizionario è una struttura dati generica
Non ha un ordine garantito (anche se in Python moderno mantiene l’ordine di inserimento).
Le chiavi possono essere di qualunque tipo.
Serve solo per mappare chiave → valore.
Non ha metodi statistici o analitici.
È un contenitore “statico”.

2-Una Series è una struttura dati NUMERICA e ANALITICA
Una Series è:ordinata,indicizzata,vettoriale (si comporta come un array)
ottimizzata per calcoli numerici,dotata di metodi statistici integrati
È pensata per l’analisi dei dati.

Cosa posso fare con una Series che NON puoi fare con un dizionario?
1. Operazioni matematiche vettoriali
python
fatturati_series * 2
Restituisce una nuova Series con tutti i valori raddoppiati.

Con un dizionario dovresti fare un ciclo.

2. Statistiche immediate
fatturati_series.mean()
fatturati_series.max()
fatturati_series.min()
fatturati_series.std()
Con un dizionario: impossibile senza scrivere codice aggiuntivo.
3. Filtri logici
fatturati_series[fatturati_series > 20000]
Restituisce solo gli anni con fatturato > 20.000.

Con un dizionario: devi fare un ciclo + condizione.

4. Allineamento automatico degli indici
Se sommi due Series con indici diversi, Pandas allinea automaticamente gli anni.

Con un dizionario: devi gestire tu le chiavi mancanti.

5. Conversione immediata in DataFrame
fatturati_series.to_frame()
Con un dizionario: devi costruire tu il DataFrame.
6. Plot immediato
fatturati_series.plot()
Con un dizionario: devi convertirlo prima.
Un plot è una visualizzazione grafica dei dati.Serve per: vedere andamenti,confrontare valori
individuare picchi e cali e capire trend nel tempo.

Esercizio L'azienda Object SpA ha un dataset con tutti gli stipendi dei dipendenti, memorizzato in una Series: import pandas as pd stipendi = pd.Series( [100, 200, 300, 400, 500,  600, 700, 800, 900, 1000] ) L'azienda ci chiede di raddoppiare tutti gli stipendi; facciamolo in due modi: • con un ciclo for e con il masking

```python
import pandas as pd

stipendi = pd.Series([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
```

```python
stipendi_doppi_for = []

for valore in stipendi:
    stipendi_doppi_for.append(valore * 2)

stipendi_doppi_for = pd.Series(stipendi_doppi_for)
print(stipendi_doppi_for)
```

Raddoppiare gli stipendi con il masking (metodo Pandas)
In realtà qui non serve un vero “filtro”: Pandas permette operazioni vettoriali direttamente sulla Series.

```python
stipendi_doppi_masking = stipendi * 2
print(stipendi_doppi_masking)
```

Esercizio Scrivere un programma che, data una lista di numeri, fornisca in output i tre numeri più grandi; gestire il caso in cui la lista sia più corta di tre, e quando uno o più dei numeri selezionati sono uguali.

```python
numeri = [10, 50, 20, 50, 30, 40]

# Se la lista ha meno di 3 elementi, lo segnaliamo
if len(numeri) < 3:
    print("La lista ha meno di tre numeri:", numeri)
else:
    # Ordiniamo la lista in ordine decrescente
    ordinati = sorted(numeri, reverse=True)

    # Prendiamo i primi tre
    top3 = ordinati[:3]

    print("I tre numeri più grandi sono:", top3)
```

```python
numeri = [10, 50, 20, 50, 30, 40]

if len(numeri) < 3:
    print("La lista ha meno di tre numeri:", numeri)
else:
    # Inizializziamo i tre massimi
    max1 = max2 = max3 = float('-inf')

    for n in numeri:
        if n > max1:
            max3 = max2
            max2 = max1
            max1 = n
        elif n > max2:
            max3 = max2
            max2 = n
        elif n > max3:
            max3 = n

    print("I tre numeri più grandi sono:", [max1, max2, max3])
```

Esercizio Scrivere un programma che • in input acquisisce una lista di numeri e un numero K • in output, dovrà restituire la media di tutti i numeri nella lista maggiori o uguali a K • se non ce ne dovesse essere nessuno, dovrà stampare a schermo un messaggio adeguato.

```python
# Chiediamo di inserire una lista di numeri.
# eval() converte la stringa inserita in una vera lista Python.
lista = eval(input("Inserisci una lista di numeri: "))

# Chiediamo il valore di K, che useremo come soglia.
K = float(input("Inserisci il valore di K: "))

# Creiamo una nuova lista contenente solo i numeri >= K.
# Questa è una list comprehension: scorre la lista e tiene solo i valori validi.
numeri_validi = [n for n in lista if n >= K]

# Se la lista filtrata è vuota, significa che nessun numero soddisfa la condizione.
if len(numeri_validi) == 0:
    print("Nessun numero è maggiore o uguale a K.")
else:
    # Calcoliamo la media: somma dei numeri diviso il loro conteggio.
    media = sum(numeri_validi) / len(numeri_validi)

    # Stampiamo il risultato.
    print("La media dei numeri >= K è:", media)
```

```python
lista_cf = [
    "ABCDEF95G01A123B",
    "GHIJKL91M02A321C",
    "MNOPQR89S03A456D",
    "STUVWX95Z04A654E",
    "XYZABC01D05A789F",
    "DEFGHI95J06A987G"
]

# Lista vuota dove inseriremo i codici fiscali che contengono "95"
cf_con_95 = []

# Scorriamo la lista
for cf in lista_cf:
    # Se contiene "95", lo aggiungiamo alla lista
    if "95" in cf:
        cf_con_95.append(cf)

# Stampiamo la lista finale
print("Codici fiscali che contengono '95':")
print(cf_con_95)

# Ora estraiamo nome e cognome per ciascuno
print("\nDettagli (cognome e nome):")
for cf in cf_con_95:
    cognome = cf[0:3]   # primi 3 caratteri
    nome = cf[3:6]      # successivi 3 caratteri
    print(f"{cf} → Cognome: {cognome}, Nome: {nome}")
```

Versione con list comprehension

```python
lista_cf = [
    "ABCDEF95G01A123B",
    "GHIJKL91M02A321C",
    "MNOPQR89S03A456D",
    "STUVWX95Z04A654E",
    "XYZABC01D05A789F",
    "DEFGHI95J06A987G"
]

# Usiamo una list comprehension per filtrare i codici fiscali che contengono "95".
# Per ogni cf in lista_cf, lo includiamo solo se "95" è presente nella stringa.
cf_con_95 = [cf for cf in lista_cf if "95" in cf]

# Stampiamo la lista risultante.
print("Codici fiscali che contengono '95':")
print(cf_con_95)

# Ora estraiamo cognome e nome per ciascun codice fiscale filtrato.
print("\nDettagli (cognome e nome):")
for cf in cf_con_95:
    # I primi 3 caratteri rappresentano il cognome.
    cognome = cf[0:3]
    # I successivi 3 caratteri rappresentano il nome.
    nome = cf[3:6]
    # Stampiamo il risultato in modo leggibile.
    print(f"{cf} → Cognome: {cognome}, Nome: {nome}")
```

Esercizio Abbiamo una lista di studenti: studenti = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry", "Isabelle", "John"] vogliamo dividere gli studenti in due squadre per un campionato di Uno nel seguente modo: selezioneremo i nomi in posizione pari per un squadra, e i nomi in posizione dispari per l'altra. Creiamo due liste per ogni squadra, e alla fine visualizziamole.

```python
#Obiettivo Dividere gli studenti in due squadre:
#Squadra A → studenti in posizione pari (0, 2, 4, …)
#Squadra B → studenti in posizione dispari (1, 3, 5, …)
# in Python gli indici partono da 0.

studenti = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry", "Isabelle", "John"]

# Liste vuote per le due squadre
squadra_pari = []
squadra_dispari = []

# Scorriamo gli indici della lista
for i in range(len(studenti)):
    # Se l'indice è pari → squadra pari
    if i % 2 == 0:
        squadra_pari.append(studenti[i])
    # Se l'indice è dispari → squadra dispari
    else:
        squadra_dispari.append(studenti[i])

# Stampiamo i risultati
print("Squadra (posizioni pari):", squadra_pari)
print("Squadra (posizioni dispari):", squadra_dispari)
```

Esercizio Abbiamo una lista con i guadagni degli ultimi 12 mesi (supponiamo da Gennaio a Dicembre): guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50] dobbiamo confrontare, stampando tutto a video, il guadagno di ogni mese con la media dei guadagni precedenti, e specificare nell'output se il guadagno attuale è maggiore o minore della media dei precedenti. Esempio di un possibile output: Mese 1: 100 € Mese 2: 90 € (media prec: 100 € - il guadagno attuale è minore) Mese 3: 70 € (media prec: 95 € - il guadagno attuale è minore)

```python
guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

# Variabile per accumulare la somma dei guadagni precedenti
somma_precedenti = 0

# Cicliamo su tutti i mesi usando l'indice
for i in range(len(guadagni)):
    
    # Guadagno del mese corrente
    guadagno_attuale = guadagni[i]
    
    # Caso speciale: il primo mese non ha precedenti
    if i == 0:
        print(f"Mese {i+1}: {guadagno_attuale} €")
    
    else:
        # Calcoliamo la media dei guadagni precedenti
        media_precedenti = somma_precedenti / i
        
        # Confrontiamo il guadagno attuale con la media
        if guadagno_attuale > media_precedenti:
            confronto = "maggiore"
        else:
            confronto = "minore"
        
        # Stampiamo il risultato
        print(f"Mese {i+1}: {guadagno_attuale} € "
              f"(media prec: {media_precedenti:.0f} € - il guadagno attuale è {confronto})")
    
    # Aggiorniamo la somma dei precedenti aggiungendo il guadagno attuale
    somma_precedenti += guadagno_attuale
```

Soluzione con Pandas

```python
import pandas as pd

# Creiamo la Series dei guadagni
guadagni = pd.Series([100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50])

# Cicliamo sugli indici della Series
for i in range(len(guadagni)):
    
    guadagno_attuale = guadagni[i]
    
    # Caso speciale: il primo mese non ha precedenti
    if i == 0:
        print(f"Mese {i+1}: {guadagno_attuale} €")
    
    else:
        # Media dei guadagni precedenti usando slicing Pandas
        media_precedenti = guadagni[:i].mean()
        
        # Confronto
        confronto = "maggiore" if guadagno_attuale > media_precedenti else "minore"
        
        print(
            f"Mese {i+1}: {guadagno_attuale} € "
            f"(media prec: {media_precedenti:.0f} € - il guadagno attuale è {confronto})"
        )
```
